```python
import pandas as pd
import numpy as np
import scipy.optimize as opt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("breast cancer kaggle.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
feature_df = data[['concave points_mean', 'fractal_dimension_mean', 'smoothness_mean', 'texture_mean']]
X = np.asanyarray(feature_df)
```

```python
data.head(30)
```

```python
data['diagnosis'] = data['diagnosis'].astype('object')
y = np.asanyarray(data['diagnosis'])
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
print("Train Set: ", x_tr.shape, y_tr.shape)
print("Test Set: ", x_te.shape, y_te.shape)
```

```python
from sklearn import svm
```

```python
sm = svm.SVC(kernel='rbf')
sm.fit(x_tr, y_tr)
```

```python
yhat = sm.predict(x_te)
yhat[0:5]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
```

```python
from sklearn.metrics import f1_score, jaccard_score
```

```python
print('F1 Score: ', f1_score(y_te, yhat, average='weighted'))
print('Jaccard Score: ', jaccard_score(y_te, yhat, pos_label='B'))
```
